```python
import pandas as pd
```

```python
jan_df = pd.read_csv('Resources/JC-20161-citibike-tripdata.csv')
feb_df = pd.read_csv('Resources/JC-20162-citibike-tripdata.csv')
mar_df = pd.read_csv('Resources/JC-20163-citibike-tripdata.csv')
apr_df = pd.read_csv('Resources/JC-201604-citibike-tripdata.csv')
may_df = pd.read_csv('Resources/JC-201605-citibike-tripdata.csv')
june_df = pd.read_csv('Resources/JC-201606-citibike-tripdata.csv')
july_df = pd.read_csv('Resources/JC-201607-citibike-tripdata.csv')
aug_df = pd.read_csv('Resources/JC-201608-citibike-tripdata.csv')
sep_df = pd.read_csv('Resources/JC-201609-citibike-tripdata.csv')
oct_df = pd.read_csv('Resources/JC-201610-citibike-tripdata.csv')
nov_df = pd.read_csv('Resources/JC-201611-citibike-tripdata.csv')
dec_df = pd.read_csv('Resources/JC-201612-citibike-tripdata.csv')
```

```python
# concactenate the df to one
citi_df = pd.concat([jan_df, feb_df, mar_df, apr_df, may_df, june_df, july_df, aug_df, sep_df, oct_df, nov_df, dec_df])
citi_df.head()
```

```python
# Creating new column to replace "Trip Duration" with it being measured in minutes instead of seconds.
citi_df["Trip_Dur_Min"] = (citi_df["Trip Duration"]/60).astype(int)
```

```python
citi_df.head()
```

```python
# Create a date column
# split date from time using .split()
citi_df[['Date','Start_Time']] = citi_df["Start Time"].str.split(' ', expand=True)
citi_df[['Stop_Date', 'Stop_Time']] = citi_df['Stop Time'].str.split(' ', expand=True)
```

```python
citi_df.info()
```

```python
m = {1 : "Male", 2 : "Female", 0: "Unknown"}
citi_df['Gender'] = citi_df['Gender'].map(m)
```

```python
citi_df.head()
```

```python
citi_df = citi_df.drop(columns=['Start_Time','Stop_Date', 'Stop_Time'])
```

```python
citi_df.head()
```

```python
# save combined data to new csv
citi_df.to_csv("Citi2016.csv")
```
