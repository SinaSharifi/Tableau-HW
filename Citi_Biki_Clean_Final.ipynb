```python
import pandas as pd
import numpy as np
import time
```

```python
citi_data = "Resources/JC-201702-citibike-dirtytripdata.csv"

citi_data2 = "Resources/JC-201703-citibike-tripdata.csv"
```

```python
citi_data_df = pd.read_csv(citi_data)
citi_data_df.head()
```

```python
citi_data2_df = pd.read_csv(citi_data2)
citi_data2_df.head()
```

```python
# Creating new column to replace "Trip Duration" with it being measured in minutes instead of seconds.
citi_data_df["Trip_Dur_Min"] = (citi_data_df["Trip Duration"]/60).astype(int)
```

```python
citi_data_df.head()
```

```python
# Create a date column
# split date from time using .split()
citi_data_df[['Date','Start_Time']] = citi_data_df["Start Time"].str.split(' ', expand=True)
citi_data_df[['Stop_Date', 'Stop_Time']] = citi_data_df['Stop Time'].str.split(' ', expand=True)
```

```python
citi_data_df.info()
```

```python
m = {1 : "Male", 2 : "Female", 0: "Unknown"}
citi_data_df['Gender'] = citi_data_df['Gender'].map(m)
```

```python
citi_data_df.head()
```

```python
citidata_feb = citi_data_df.drop(columns=['Trip Duration', 'Start Time', 'Stop Time', 'Bike ID', 'Birth Year', 'Start_Time', 'Stop_Date', 'Stop_Time'])
```

```python
citidata_feb.head()
```

```python
citidata_feb.to_csv("Citibike-Feb2017.csv")
```

```python
# Experiment with the clean data
citidata_feb["Trip_Dur_Min"].mean()
```

```python
citidata_feb["User Type"].value_counts()
```

```python
citidata_feb["Gender"].value_counts()
```
